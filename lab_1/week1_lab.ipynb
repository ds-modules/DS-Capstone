### Question 4

You realize that the `cups` column has different cups per serving, so that the nutritional information is not standardized across cereals.
Since you normally eat 1 cup of cereal for breakfast, it would be better to see the nutritional values for each cereal per cup. 

Let's transform the data so that the dataframe reflects the nutritional values of each cereal per cup and store it in a new table called `serving`. `serving` should have all the columns of `cereal_drop`, but the `'calories', 'protein', 'fiber', 'sugars'` columns should reflect serving per 1 cup. 

*Hint 1:* For example, the serving size for cheerios is 1.25 and the caloric content is 110. We can divide calories by cups (110 / 1.25) to get calories per cup. Similarly, All-Bran with Extra Fiber has 50 calories for 0.5 (or 1/2) cups of serving. Could we divide calories by cups here as well?

As a challenge, you can try using `df.div`. You can find the documentation [here](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.divide.html). Don't forget the `axis` argument.

```
BEGIN QUESTION
name: q4
```

````python
""" ## BEGIN PROMPT
serving = ...
serving

## END PROMPT """

# BEGIN SOLUTION NO PROMPT
## SOLUTION ##
serving = cereal[['name', 'protein', 'fiber', 'calories', 'sugars', 'cups']]
serving['protein'] = serving['protein'] / serving['cups']
serving['fiber'] = serving['fiber'] / serving['cups']
serving['calories'] = serving['calories'] / serving['cups']
serving['sugars'] = serving['sugars'] / serving['cups']
serving = serving.drop('cups', axis = 1)
serving
# END SOLUTION
````

## Question 8 - Melt

What if you receive data that is in the format of a pivot table, like the one above?  Sometimes we want to reverse the effects of a pivot, and that's when the pd.melt() function comes in handy.  Below, we have provided df2, which is in the format of a pivoted table.

````python
df2 = pd.DataFrame({'SubjectID': [1000, 1001, 1002], 'Before Meds': [4.5, 2.9, 3.3], 'After Meds': [4.8, 3.8, 4.2]})

df2
````

In this case, the data in df2 will be a lot easier to analyze if it was presented in this form.


<img src="solution.png" /> 

Using pd.melt(), assign melt_df to a DataFrame that looks like the above image.  The documentation for df.melt() can be found [here](https://pandas.pydata.org/docs/reference/api/pandas.melt.html).  If you have any trouble figuring out this question, please visit us in office hours. 


````python
## SOLUTION ##
melt_df = pd.melt(df2, id_vars = ['SubjectID'], value_vars = ['Before Meds', 'After Meds'], var_name = 'TimePoint', value_name = 'Response Score') # SOLUTION
melt_df
````

````python
## Test ##

assert all(melt_df.columns == ['SubjectID', 'TimePoint', 'Response Score'])
````

````python
## Test ##

assert all(melt_df['SubjectID'] == np.array([1000, 1001, 1002, 1000, 1001, 1002]))
````

## Question 9 

Currently, the cereal table has a numerical column called ratings.  What if we wanted to create another column that contains the categorial quantiles for each cereal's ratings?  For example, if the rating for Almonds Delight is within the highest 25% of the ratings, then the new column 'Rating Quartiles' for this cereal would contain "Q4".  The first few rows of the table should look like this: 

<img src="quartiles.png" /> 

Add a new column to the cereal table called "Rating Quartiles" that gives the quartile for the cereal's rating. 

Note: There is a Pandas function that you can use to do this question; you don't have to manually assign the quartiles or use a user-defined function with df.apply().  You can use user-defined functions, but we strongly encourage you to look up how to create the quartiles in Pandas.

````python
## SOLUTION ##

cereal['Rating Quartiles'] = pd.qcut(cereal['rating'], q = 4, labels = ['Q1', 'Q2', 'Q3', 'Q4'])#SOLUTION
cereal.head() 
````
